# 定义类和对象

# 类结构  类名 属性 方法

class 类名:

    属性

    方法

```python
class Person:
    '''
    对应人的特征
    '''
    # name='zhaojianhe' #类属性
    age=28 #类属性：  在类的内部定义的变量
    '''
    对应人的行为
    '''
    # 实例方法： 在类的内部，使用def关键字可以定义，与一般函数定义不同，类方法必须包含参数self，且为第一个参数（名字或者标识可以是其他，单位制必须被占用）
    # 实例方法是归于 类的实例所有
    # 属性
    # 定义在方法里面使用self引用的属性称之为实例属性
    def __init__(self):
        self.name='zhaojianhe' #实例属性,归实例对象所有
        pass
        
    def eat(self):
        print('大口吃饭') 
        pass
    def run(self):
        print('快速跑')
        pass
    pass
```

# 创建对象规则格式 对象名=类名()

```python
#创建对象，对象名=类（），相当于类的实例化
xm=Person()
xm.age#类的属性
xm.eat()#调用函数
print('{}的年龄：{}'.format(xm.name,xm.age))
```

```python
#创建实例对象
xw=Person()
xw.eat() #实例方法调用
```
