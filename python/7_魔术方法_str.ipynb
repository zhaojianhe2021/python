```python
class people:
    
    def __init__(self,name,sex,age):#python中已经内置好的：魔术方法
        '''
        实例属性的传参
        '''
        self.name=name#给一个初始值
        self.sex=sex
        self.age=age
        pass
    def __str__(self):
        return('%s性别是%s，今年%s岁了'%(self.name,self.sex,self.age))
    
    
    def eat(self,food):
        '''
        吃的行为
        return:
        '''
        print(self.name+'喜欢吃'+food)
    pass
```

```python
xm=people('xiaoming','nv',18)
print(xm)
```
