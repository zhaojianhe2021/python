```python
import psutil
```

```python
def get_cpu_info():
    cpu1 = psutil.cpu_count()
    print("cpu逻辑个数:",cpu1)
```

```python
def get_mem_info():
    mem = psutil.virtual_memory()
    mem1 = str(mem.total/1024/1024/1024)
    mem2 = str(mem.free/1024/1024/1024)
    print("内存总数为:",mem1[0:3],"G")
    print("空闲内存总数:", mem2[0:3], "G")
```

```python
get_cpu_info()
get_mem_info()
```

# 全局变量和局部变量

```python
pro='工科'#全局变量：作用域在所有
name='teacher'# 当全局变量和局部变量出现重复定义的时候，程序会优先使用函数内部的变量（强龙不压地头蛇）
def printInfo():
    name='peter'# 局部变量：在函数内部定义的变量，作用域仅仅局限在函数内部
    print('{},{}'.format(name,pro))
    pass
def TestMethod():
#     name='omom'
    print(name)
    pass

def changeGloble():#在函数内部修改全局变量，使用global关键字声明
    global pro
    pro='市场影响'
    pass

changeGloble()
print(pro)
TestMethod()
printInfo()
```
