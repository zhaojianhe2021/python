```python
import psutil
```

```python
def get_cpu_info():
    cpu1 = psutil.cpu_count()
    print("cpu:",cpu1)
```

```python
def get_mem_info():
    mem = psutil.virtual_memory()
    mem1 = str(mem.total/1024/1024/1024)
    mem2 = str(mem.free/1024/1024/1024)
    print(":",mem1[0:3],"G")
    print(":", mem2[0:3], "G")
```

```python
get_cpu_info()
get_mem_info()
```
