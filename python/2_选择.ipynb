    0:石头 1:剪刀 2:布 

```python
import random
```

```python
#加个循环
count=1
while count<=10:
    person=int(input('请出拳：[0:石头 1:剪刀 2:布]'))
    computer=random.randint(0,2)
    if person==0 and computer==1:
        print('厉害了，你赢了')
        pass
    elif person==1 and computer==2:
        print('厉害了，你赢了')
        pass
    elif person==2 and computer==0:
        print('厉害了，你赢了')
        pass
    elif person==computer:
        print('不错，竟然是平手')
        pass
    else:
        print('输了吧')
    count+=1
```
