```python
class person:
       '''
       定义类
       '''
    def __init__(sdlf,pro):
        '''
        专业
        '''
        self.pro=pro
        pass
    
 
    def eat(self,name,food):
        '''
        实例方法
        '''
        print('%s 喜欢吃 %s '%(name,food))
        pass
    
```

```python
xw=person()
print('xw=%s',id(xw))
xw.eat('小王','榴莲')
#内存地址相同：实例化对象和self指向同一个内存，说明self就是实例对象本身
```

```python
小结：self特点

self只有在定义类方法的时候才有意义，在调用时候不必传入相应参数，而是由解释器自动的去指向

self名称是可以更改的
```
