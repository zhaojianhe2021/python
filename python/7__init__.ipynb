```python
class Animal(object):
    def eat(self):
        print('吃饭')
```

```python
dog = Animal()  # 创建一个对象
dog.colour = 'red'  # 添加属性
dog.name = '旺财'  # 添加属性
print(dog.name)
```

```python
cat = Animal() # 创建一个对象
cat.colour = 'black' # 添加属性
cat.name = '小花' # 添加属性
print(cat.name)
```

```python
class people:
    def eat(self):
        '''
        吃的行为
        return:
        '''
        print('喜欢吃pizza')
    pass
```

```python
xq=people() #创建出来的对象
xq.name='小倩'#手动添加属性(实例属性)
xq.sex='女生'#手动添加属性
xq.eat()
print(xq.name,xq.sex)
```

# 如果有n个对象，需要反复添加属性，麻烦——>使用init

```python
class people:
    def __init__(self):#python中已经内置好的：魔术方法
        '''
        实例属性的统一声明
        '''
        self.name='小倩'#给一个初始值
        self.sex='女生'
        self.age='20'
        pass
    def eat(self):
        '''
        吃的行为
        return:
        '''
        print('喜欢吃pizza')
    pass
```

```python
xm=people()#创建新对象的时候，自动执行调用
print(xm.name,xm.sex,xm.age)#直接输出的是默认值
xm.name='小明'
print(xm.name,xm.sex,xm.age)#输出修改后的值
```

# init传递参数 改进

```python
class people:
    def __init__(self,name,sex,age):#python中已经内置好的：魔术方法
        '''
        实例属性的传参
        '''
        self.name=name#给一个初始值
        self.sex=sex
        self.age=age
        pass
    def eat(self,food):
        '''
        吃的行为
        return:
        '''
        print(self.name+'喜欢吃'+food)
    pass
```

```python
zp=people('张鹏','男生',18)
print(zp.name,zp.age)
zp.eat('香蕉')
```

```python
lh=people('李辉','女生',28)
print(lh.name)
lh.eat('苹果')
```

总结 __init__:

    1.python 自带内置函数 具有特殊的函数 使用双下划线包起来的【魔术方法】
    
    2.可以认为是一个初始化的方法 用来定义实例属性和初始化数据的 并且创建对象的时候自动调用 不用手动去调用
    
    3.利用传参的机制，可以让定义功能，更加强大并且方便的类
