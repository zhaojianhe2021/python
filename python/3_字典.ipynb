使用键访问数据，效率非常高，和list一样支持对数据添加、修改、删除

特点：

1：不是序列类型，没有下标概念，是一个无序的键值集合，是内置的高级数据类型

2：用{ }来表示字典对象，每个键值对用逗号分割

3：键必须是不可变的类型【元组、字符串】值可以是任意的类型

4：每个键必定是唯一的，如果重复，后者会覆盖

```python
#如何创建字典（一个键对应一个值）
dictA={'pro':'工学专业','school':'chung·ang university'}
print(type(dictA))
dictA['name']='赵建贺'
dictA['age']='28'
print(dictA)
print(len(dictA))#以逗号分割为一个项，每一项代表一个键值对
```

```python
print(dictA['name'])#通过键获取对应的值
dictA['name']='谢霆锋'#修改键对应的值
dictA['school']='香港大学'
print(dictA)
```

```python
#获取所有的键
print(dictA.keys())
#获取所有的值
print(dictA.values())
#获取所有的数据项
print(dictA.items())
```

```python
#取键值对
for item in dictA.items():
    print(item)
    pass
for key,value in dictA.items():
    print('%s==%s'%(key,value))
```

```python
#更改值(没有的话就添加)
dictA.update({'height':1.80})
print(dictA)
#删除操作1
del dictA['name']
print(dictA)
```

```python
#删除操作2
dictA.pop('age')
print(dictA)
```

```python
#排序, 按照key排序
sorted(dictA.items(),key=lambda d:d[0])
```

```python
#排序, 按照value排序
sorted(dictA.items(),key=lambda d:d[1])#类型不一致，五排排列
```
