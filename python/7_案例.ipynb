```python
import time
```

```python
class hero:
    def __init__(self,name,hp):
        '''
        构造初始化函数
        param name：角色名
        param hp：血量
        '''
        self.name=name
        self.hp=hp
        pass
    def tong(self,enemy):
        enemy.hp -= 10
        info='%s捅了%s一刀'%(self.name,enemy.name)
        print(info)
        pass
    def kan(self,enemy):
        enemy.hp -= 15
        info='%s砍了%s一刀'%(self.name,enemy.name)
        print(info)
        pass
    def chiyao(self):
        self.hp += 10
        info='%s吃了补血药'%(self.name)
        print(info)
        pass
    def __str__(self):
        return '%s还剩%s的血量'%(self.name,self.hp)
```

```python
xmcx=hero('西门吹雪',100)
ygc=hero('叶孤城',100)
while True:
    if xmcx.hp<=0 or ygc.hp<=0:
        #满足条件，退出循环
        break
        pass
    
    xmcx.tong(ygc)
    print(ygc)
    print(xmcx)
    print('--------------------------------------------------------------------------------------------')
    ygc.chiyao()
    print(ygc)
    print(xmcx)
    print('--------------------------------------------------------------------------------------------')
    ygc.kan(xmcx)
    print(ygc)
    print(xmcx)
    time.sleep(1)
print('对战结束')
```
