```python
import time
```

```python
class hero:
    def __init__(self,name,hp):
        '''
        
        param name
        param hp
        '''
        self.name=name
        self.hp=hp
        pass
    def tong(self,enemy):
        enemy.hp -= 10
        info='%s%s'%(self.name,enemy.name)
        print(info)
        pass
    def kan(self,enemy):
        enemy.hp -= 15
        info='%s%s'%(self.name,enemy.name)
        print(info)
        pass
    def chiyao(self):
        self.hp += 10
        info='%s'%(self.name)
        print(info)
        pass
    def __str__(self):
        return '%s%s'%(self.name,self.hp)
```

```python
xmcx=hero('',100)
ygc=hero('',100)
while True:
    if xmcx.hp<=0 or ygc.hp<=0:
        #
        break
        pass
    
    xmcx.tong(ygc)
    print(ygc)
    print(xmcx)
    print('--------------------------------------------------------------------------------------------')
    ygc.chiyao()
    print(ygc)
    print(xmcx)
    print('--------------------------------------------------------------------------------------------')
    ygc.kan(xmcx)
    print(ygc)
    print(xmcx)
    time.sleep(1)
print('')
```
