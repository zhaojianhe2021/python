# 语法特点：遍历操作，依次取集合容器中的每个值

for 临时变量 in 容器：
    
    执行代码块

```python
#'遍历字符串集合'

tags='我是一个中国人'
for item in tags:
    print(item,end='')
    pass
```

range 生成一个数据集合列表

range(起始值：结束值：步长)

```python
type(range(1,101))
for data in range(1,101):#z左包含，右不包含
    print(data,end=' ')
```

```python
#求和
sum=0
for data in range(1,101):
    sum+=data
print('sum=%d'%sum)
```

```python
#求偶数
for data in range(50,201):
    if data%2==0:
        print('%d是偶数'%data)
        pass
    else:
        print('%d是奇数'%data)
```

# break：中断，结束

```python
sum=0
for item in range(1,51):
    if sum>100:
        break
        pass
    sum+=item
print(sum)
    
```

# continue：结束本次循环，接续进行下次循环

```python
for item in range(1,100):
    if item%2==0:
        continue
        pass
print(item)
```

# for-----else

```python
for item in range(1,11):
    print(item,end=' ')
    if item>=5:
        break
    pass
else:
    print('已经执行完了吗')
```

# while-----else

```python
index=1
while index<=10:
    print(index)
    if index==6:
        break
    index+=1
    pass
else:
    print('else执行了吗')
```
