# list：python中非常重要的数据结构，是一种有序的数据集合

1.支持增删改查

2.列表中的数据是可以变化的【数据项可以变化，内存地址不会改变】

3.用[ ]来表示列表类型，数据项之间用逗号来分割 ，数据项可以是任何数据类型

4.支持索引和切片来进行操作

```python
li=[]#空列表
print(type(li))
li=[1,2,3,'你好']
print(len(li))# len函数可以获取到列表中的数据个数
```

# 查找数据

```python
listA=['abcd',785,12.23,'qiuzhi',True]
print(listA) # 输出完整的列表 
print(listA[0]) #输出第一个元素
print(listA[1:3])
print(listA*3)
```

# 增加数据

```python
listA=['abcd',785,12.23,'qiuzhi',True]
print(listA)
listA.append(['fff','ddd']) # 追加列表-----------append
print(listA)
listA.append(8888)
print(listA)
listA.insert(1,'在指定的位置插入') # 插入列表-----------insert
print(listA)
rsData=list(range(10)) # 强制转换为list对象
listA.extend(rsData) # 扩展 等于批量添加-----------extend
print(listA)
listA.extend([11,22,33,44])
print(listA)
```

# 修改

```python
listA=['abcd',785,12.23,'qiuzhi',True]
listA[0]='paper'
listA[1]=52.2
print(listA)
```

# 删除

```python
listB=list(range(10,20))
print(listB)
del listB[0]# 删除列表中第一个元素
print(listB)
del listB[1:3]#删除部分
print(listB)
listB.remove(14)#移除指定值的元素
print(listB)
listB.pop(0) # 移除指定索引的元素
print(listB)
```
