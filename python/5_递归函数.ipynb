```python
#
def jiecheng(n):
    result=1
    for item in range(1,n+1):
        result*=item
    return result
```

```python
print('n{}'.format(jiecheng(5)))
```

```python
def diguijc(n):
    if n==1:
        return 1
    else:
        return n*diguijc(n-1)
```

```python
diguijc(5)
```
